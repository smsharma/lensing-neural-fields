```python
xx = np.linspace(0, 1, 20)
xx, yy = np.meshgrid(xx, xx)

plt.figure(figsize=(3.5, 3.5))

plt.scatter(xx, yy, color='grey', alpha=0.5, s=2.5)

plt.xlabel(r"$\beta_x$", fontsize=25)
plt.ylabel(r"$\beta_y$", fontsize=25)

ax = plt.gca()
# ax.grid(which='both', )
# ax.grid(which='minor', alpha=0.2)
# ax.grid(which='major', alpha=0.5)

ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

plt.tight_layout()
plt.savefig("figures/coords.pdf", transparent=True)
```

```python
plt.figure(figsize=(6 / 1.5, 4 / 1.5))

x = np.linspace(0, 1, 500)
xx = np.linspace(0, 1, 10)

for i in range(4):
    plt.plot(x, np.sin(2 ** i * np.pi * x), alpha=0.5)
    plt.scatter(xx, np.sin(2 ** i * np.pi * xx))
    
for i in xx:
    plt.axvline(i, color='grey', ls='dotted', lw=0.9)
    
plt.xlabel(r"$\beta$", fontsize=24)
plt.ylabel(r"${\gamma(\beta)}$", fontsize=24)

plt.tick_params(
    axis='both',
    which='both',
    bottom=False,
    top=False,
    left=False,
    right=False,
    labelleft=False,
    labelbottom=False)

plt.tight_layout()
plt.savefig("figures/encodings.pdf", transparent=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors, colorbar

vals = np.linspace(0,1,int(64 / 3))

np.random.shuffle(vals)

cmap = plt.cm.colors.ListedColormap(plt.cm.RdBu(vals))

fig, ax = plt.subplots(1, 1, figsize=(15 / 3., 0.5))

cb = colorbar.ColorbarBase(ax, cmap=cmap, spacing='proportional', ticks=None
                                   , format='%1i', orientation=u'horizontal', alpha=0.7)

plt.axis('off')
plt.savefig("figures/fc1.pdf")
```
